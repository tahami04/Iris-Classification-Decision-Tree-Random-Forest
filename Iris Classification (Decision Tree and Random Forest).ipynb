# Task(A)

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import tree

# Reading the Iris.csv file
iris = load_iris()

# Extracting Attributes / Features
X = iris.data

# Extracting Target / Class Labels
y = iris.target

# Import Library for splitting data
from sklearn.model_selection import train_test_split

# Creating Train and Test datasets
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 50, test_size = 0.25)

# Creating Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
print(clf.fit(X_train,y_train))

# Predicting the Test set results
y_pred = clf.predict(X_test)

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)

# Making the Confusion Matrix
print ("Model Accuracy Decision Tree: ", accuracy_score(y_test, y_pred))
print("Confusion Matrix Decision Tree: \n", metrics.confusion_matrix(y_test, y_pred))
```

# Task (B)

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree

# Reading the Iris.csv file
iris = load_iris()

# Extracting Attributes / Features
X = iris.data

# Extracting Target / Class Labels
y = iris.target

# Import Library for splitting data
from sklearn.model_selection import train_test_split

# Creating Train and Test datasets
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 50, test_size = 0.25)

#Applying Rndom Forest
from sklearn.ensemble import RandomForestClassifier
# set model, fit and predict
rfc = RandomForestClassifier(max_depth=10, random_state=0)
print(rfc.fit(X_train, y_train))

# Predicting the Test set results
y_prediction = rfc.predict(X_test)

fig = plt.figure(figsize=(25, 20))
plot_tree(rfc.estimators_[0], 
          feature_names=iris.feature_names,
          class_names=iris.target_names, 
          rounded=True)

# Making the Confusion Matrix
print ("Model Accuracy Rndom Forest: ", accuracy_score(y_test, y_prediction))
print("Confusion Matrix Rndom Forest: \n", metrics.confusion_matrix(y_test, y_prediction))
```
